```python
df = final.copy()

df['discount_factor'] = 1 - (df['tot_discounts'] / df['tot_charges'])

df['price_num'] = (
    (df['ip_charges'] + df['icu_charges'] + df['ancillary_charges']) *
    df['discount_factor']
    - df['tot_mcare_payment']
)

df['price_denom'] = df['tot_discharges'] - df['mcare_discharges']

df['price'] = df['price_num'] / df['price_denom']
```

## Question 5*

```python
df = final.copy()

df['penalty'] = (
    df['hrrp_payment'].fillna(0) +
    df['hvbp_payment'].fillna(0)
)
```

```python
df['penalized'] = df['penalty'] != 0
```

```python
q5 = (
    df[(df['year'] >= 2012) & (df['year'] <= 2019)]
    .groupby('year')['penalized']
    .mean()
    .reset_index(name='share_penalized')
)
```

```python
plt.figure(figsize=(8,5))

plt.plot(q5['year'], q5['share_penalized'], marker='o')

plt.title('Share of Hospitals Penalized (HRRP/VBP), 2012–2019')
plt.xlabel('Year')
plt.ylabel('Share Penalized')

plt.ylim(0, 1)  # important for interpretation

plt.show()
```

```python
# HRRP Penalized Hospitals 
share_year = (
    df
    .assign(
    penalized=np.where(
        (df["hrrp_payment"] > 0) | (df["hvbp_payment"] > 0),
        1,
        0
    )
    )
    .groupby("year", as_index=False)
    .agg(share_penalized=("penalized", "mean"))
)

share_year
```

```python
question_5 = share_year[share_year["year"] > 2011]
fig, ax = plt.subplots()
ax.plot(question_5["year"].astype(str), question_5["share_penalized"], linewidth=1)
ax.set_xlabel("Year")
ax.set_ylabel("Share of Hospitals")
ax.set_title("Penalized Hospitals by Year")
ax.tick_params(axis="x", rotation=90)
plt.tight_layout()
plt.show()
```

## Question 6

```python
df = final.copy()

df['discount_factor'] = 1 - (df['tot_discounts'] / df['tot_charges'])

df['price_num'] = (
    (df['ip_charges'] + df['icu_charges'] + df['ancillary_charges']) *
    df['discount_factor']
    - df['tot_mcare_payment']
)

df['price_denom'] = df['tot_discharges'] - df['mcare_discharges']
df['price'] = df['price_num'] / df['price_denom']

df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna(subset=['price'])

df = df[(df['price'] > 0) & (df['price'] < 100000)].copy()

df['penalty'] = (
    df['hrrp_payment'].fillna(0) +
    df['hvbp_payment'].fillna(0)
).abs()

df['penalty_k'] = df['penalty'] / 1000

price_2011 = (
    df[df['year'] == 2011][['provider_number', 'price']]
    .rename(columns={'price': 'price_2011'})
)

price_2014 = (
    df[df['year'] == 2014][['provider_number', 'price']]
    .rename(columns={'price': 'price_2014'})
)

price_change = price_2011.merge(price_2014, on='provider_number', how='inner')
price_change['price_change'] = price_change['price_2014'] - price_change['price_2011']

penalty_2012 = (
    df[df['year'] == 2012][['provider_number', 'penalty_k']]
    .drop_duplicates(subset=['provider_number'])
)

pre = df[(df['year'] >= 2009) & (df['year'] <= 2011)].copy()

pre_avg = (
    pre.groupby('provider_number', as_index=False)
    .agg({
        'beds': 'mean',
        'mcaid_discharges': 'mean',
        'mcare_discharges': 'mean'
    })
)

reg = price_change.merge(penalty_2012, on='provider_number', how='inner')
reg = reg.merge(pre_avg, on='provider_number', how='inner')

reg = reg.dropna(subset=[
    'price_change', 'penalty_k', 'beds', 'mcaid_discharges', 'mcare_discharges'
]).copy()

print("Regression sample shape:", reg.shape)
print(reg[['provider_number', 'price_change', 'penalty_k', 'beds', 'mcaid_discharges', 'mcare_discharges']].head())
```

```python
m1 = smf.ols('price_change ~ penalty_k', data=reg).fit()
m2 = smf.ols('price_change ~ penalty_k + beds', data=reg).fit()
m3 = smf.ols('price_change ~ penalty_k + beds + mcaid_discharges', data=reg).fit()

q6_table = pd.DataFrame({
    "OLS1": [
        m1.params["penalty_k"],
        m1.bse["penalty_k"],
        int(m1.nobs)
    ],
    "OLS2": [
        m2.params["penalty_k"],
        m2.bse["penalty_k"],
        int(m2.nobs)
    ],
    "OLS3": [
        m3.params["penalty_k"],
        m3.bse["penalty_k"],
        int(m3.nobs)
    ]
},
index=[
    "Net Penalty ($1000s)",
    "Std. Error",
    "N"
])

q6_table.round(3)
```

## Question 9

```python
from linearmodels.iv import IV2SLS
```

```python
m1 = smf.ols('price_change ~ penalty_k', data=reg).fit()
m2 = smf.ols('price_change ~ penalty_k + beds', data=reg).fit()
m3 = smf.ols('price_change ~ penalty_k + beds + mcaid_discharges', data=reg).fit()
```

```python
iv1 = IV2SLS.from_formula(
    'price_change ~ 1 + [penalty_k ~ mcare_discharges]',
    data=reg
).fit()

iv2 = IV2SLS.from_formula(
    'price_change ~ beds + [penalty_k ~ mcare_discharges]',
    data=reg
).fit()

iv3 = IV2SLS.from_formula(
    'price_change ~ beds + mcaid_discharges + [penalty_k ~ mcare_discharges]',
    data=reg
).fit()
```

```python
iv_table = pd.DataFrame({
    "IV1": [
        iv1.params["penalty_k"],
        iv1.std_errors["penalty_k"],
        int(iv1.nobs)
    ],
    "IV2": [
        iv2.params["penalty_k"],
        iv2.std_errors["penalty_k"],
        int(iv2.nobs)
    ],
    "IV3": [
        iv3.params["penalty_k"],
        iv3.std_errors["penalty_k"],
        int(iv3.nobs)
    ]
},
index=[
    "Net Penalty ($1000s)",
    "Std. Error",
    "N"
])

iv_table.round(3)
```

## Question 10

The IV estimate identifies a local ATE, reflecting the effect of penalties on hospital prices for hospitals whose penalties are influenced by their pre-2012 Medicare discharges. It captures the effect for hospitals with higher Medicare exposure, as these hospitals are more affected by discharges. This local effect may differ from the overall ATE because it does not represent all hospitals, particularly those with low Medicare discharges whose penalties are less influenced by the discharges.
